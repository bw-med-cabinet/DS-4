```
df["Flavor"] = df["Flavor"].fillna("None")
```

```
total_flavor = []
max_len = 0
total_flavor.append("None")
for val in df["Flavor"].values:
    print(val)
    val = val.split(",")
    
    if len(val) > max_len:
        max_len = len(val)
    
    for f in val:
        if f not in total_flavor:
            total_flavor.append(f)
```
